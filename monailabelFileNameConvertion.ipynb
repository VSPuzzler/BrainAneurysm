```python
import os

def rename_matching_files(folder1, folder2):
    # Get the list of file names in both folders
    files_in_folder1 = os.listdir(folder1)
    files_in_folder2 = os.listdir(folder2)

    # Create a dictionary to map the first 20 characters to full file names in folder1
    file_map = {file[:20]: file for file in files_in_folder1}

    for file in files_in_folder2:
        # Check if the first 20 characters match any in folder1
        match = file_map.get(file[:20])
        if match:
            # Construct full file paths
            original_path = os.path.join(folder2, file)
            new_path = os.path.join(folder2, match)

            # Rename the file in folder2
            os.rename(original_path, new_path)
            print(f"Renamed '{file}' to '{match}'")

# Example usage
folder1 = 'path/to/folder1'  # Replace with the actual path
folder2 = 'path/to/folder2'  # Replace with the actual path
rename_matching_files(folder1, folder2)
```
